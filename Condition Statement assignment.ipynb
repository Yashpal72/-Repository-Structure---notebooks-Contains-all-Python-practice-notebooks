Q1. Write a program to check whether a number is positive or negative.

```python
# taking input Frome the user 
num = int(input("Enter a number"))

if num > 0:
    print("It is Postive")

elif num < 0:
    print("It is Negtive")

else: 
    print("Number is zero")
```

Q3 Write a program to assign grades based on marks

```python
# Taking input from the user
marks = int(input("Enter your marks: "))

# Checking grade
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
elif marks >= 35:
    print("Grade: D")
else:
    print("Grade: Fail")
```
