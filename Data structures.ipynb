# There are two type of data structures

**1. Mutable**

Mutable is when something is changeable or have the ability to change in python, 'mutable' is the ability to change vaules. These are often objects

- List[]
- sets{}
- dictionaries {key:values}

**2. Immutable**

It is not possible to alter the value of an object in python, it is known as immutable object. Once Immutable object is created it's value remains permanent and unchangeable.
- numerical (int, float)
- strings tuples
- tuples

```python
# immutable data structures
# int Float str ---> cannot modify
```

```python
a = 10
print(a)
print(id(a))
```

```python
a = 20
print(a)
print(id(a))
```
