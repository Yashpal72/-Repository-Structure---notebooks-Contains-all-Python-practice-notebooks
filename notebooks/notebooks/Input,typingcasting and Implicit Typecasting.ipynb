```python
# general example

# a = 'yash'
#int(a)

a = '10'
print(a,'\t',type(a))
print(float (a),'\t',type(a))
print(float(a),'\t',type(float(a)))
print(int(a),'\t', type(int(a)))
print(bool (a),'\t', type(bool (a)))
```

```python
# bottom to top

a = True
print(bool(a),'\t',type(bool(a)))
print(int(a),'\t',type(int(a)))
print(float(a),'\t',type(float(a)))
print(str(a),'\t',type(str(a)))
```

```python
#Calculator
print('menu')
print('-'*50)
print('1. Addition')
print('2. Subraction')
print('3. division')
print('4. Multiplication')
print('5. Exit')
choice = int(input('enter your choice(number only)'))

if choice>0 and choice<6:
    if choice==1: #nested if else 
        n1 = int(input('enter your choice'))
        n2 = int(input('enter your choice'))
        print('sum of n1 and n2 ',n1+n2)

    elif choice==2: # nested if else
        n1 = int(input('enter your choice'))
        n2 = int(input('enter your choice'))
        print('Subraction of n1 and n2 ',n1-n2)
        
    elif choice==3: # nested if else
        n1 = int(input('enter your choice'))
        n2 = int(input('enter your choice'))
        print('division of n1 and n2 ',n1/n2)

    elif choice==2: # nested if else
        n1 = int(input('enter your choice'))
        n2 = int(input('enter your choice'))
        print('Multiplication of n1 and n2 ',n1*n2)

    elif choice==5: #nested if else
        print('exit')


```
