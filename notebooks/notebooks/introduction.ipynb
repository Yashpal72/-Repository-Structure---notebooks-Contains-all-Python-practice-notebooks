# Variable:
Variables are used to to store Data Values.They act as placeholders for various types of data such as lists, numbers, strings or objects .

# rules for assigning a variables :

- Variable names must start with a letter or underscore(_), followed by letters , numbers, or underscores.
- Variable names are case sensitive
- No special Characters(except underscores) are allowed in variable names.
- Keyword cannot be used for assigning a variable like if-else, elif, while , def etc
